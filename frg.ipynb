```python
p=dfc.groupby('Country')['Quantity'].sum().nlargest(5)
plt.pie(p,labels=p.index,autopct='%1.1f%%')
plt.show()
```

```python
c=dfc.groupby('CustomerID')['Quantity'].sum().nlargest(10)
plt.pie(c,labels=c.index,autopct='%1.1f%%')
plt.show()
```

```python
plt.Figure(figsize=(10,6))

sns.barplot(dfc['StockCode'])
plt.show()
```

```python
import matplotlib.pyplot as plt

dfc["InvoiceDate"] = pd.to_datetime(dfc["InvoiceDate"])  # Ensure datetime format
dfc.set_index("InvoiceDate")["Quantity"].resample("M").sum().plot(kind="line", figsize=(12, 5))
plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Quantity Sold")
plt.show()
```

```python
import seaborn as sns

top_products = dfc.groupby("Description")["Quantity"].sum().nlargest(10)
plt.figure(figsize=(12, 5))
sns.barplot(x=top_products.values, y=top_products.index, palette="viridis")
plt.title("Top 10 Best-Selling Products")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Product Description")
plt.show()
```
